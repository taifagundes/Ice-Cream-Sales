```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IceCreamSales.csv')

df
```

```python
# Excluindo duas colunas
colunas_para_excluir = ['Date', 'Time']

df = df.drop(columns=colunas_para_excluir)

df
```

```python
# Modificando os nomes das colunas
novos_nomes = ['Temperature', 'Sales']

df.columns = novos_nomes

df
```

```python
# Classificando as primeiras 11 linhas do maior para o menor
result = df.head(11).sort_values('Temperature', ascending = False)

result
```

```python
# Descrevendo como vai ficar o grfico plt
plt.plot(result['Temperature'], result['Sales'], 'r.--' ,label = 'US$' )

plt.title('Ice Cream Truck')

plt.ylabel('Sales (US$)')
plt.xlabel('Temperature (Celsios)')


plt.legend()
plt.savefig('IceCream_sales.pdf')
plt.show()
```
